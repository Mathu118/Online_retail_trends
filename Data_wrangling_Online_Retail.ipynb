```python
# Introduce regex pattern
apparel = '|'.join(apparel)
accessories = '|'.join(accessories)
entertainment = '|'.join(entertainment)
household = '|'.join(household)
decoration = '|'.join(decoration)
food = '|'.join(food)
furniture = '|'.join(furniture)
```

```python
# Assign a group name to the correspondnig regex pattents of food items
import warnings
warnings.filterwarnings('ignore')
df2["Category"] = ""
df2["Category"][df2['Description'].str.contains(apparel)] = "Apparel"
df2["Category"][df2['Description'].str.contains(accessories)] = "Accessories"
df2["Category"][df2['Description'].str.contains(entertainment)] = "Entertainment"
df2["Category"][df2['Description'].str.contains(household)] = "Household"
df2["Category"][df2['Description'].str.contains(decoration)] = "Decoration"
df2["Category"][df2['Description'].str.contains(food)] = "Food"
df2["Category"][df2['Description'].str.contains(furniture)] = "Furniture"
```
